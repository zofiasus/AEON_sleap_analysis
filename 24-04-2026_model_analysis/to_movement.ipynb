```python
from movement.io import load_dataset
from pathlib import Path
import xarray as xr
import numpy as np
```

```python
REMOTE_PATH = Path("/Volumes/homes/live/zsus/aeon_sleap/abcEphysPilot01/")
OUT_PATH = Path(REMOTE_PATH, "processed/")
PRED_PATH = Path(REMOTE_PATH, "predictions/")
TIMESTAMP = "2026-04-14T14-00-00"

CAMERAS = ['CameraEast', 'CameraNorth', 'CameraSouth', 'CameraWest', 'CameraNest', 'CameraTop']
```

```python
def get_video_path(prediction_path: Path, camera: st, timestamp: str) -> Path:
    return prediction_path / f"{camera}_{camera}_{timestamp}.slp"
```

```python
ds_by_cam = {}
for cam in CAMERAS:
    path = get_video_path(PRED_PATH, cam, TIMESTAMP)
    ds = load_dataset(path, source_software="auto", fps=50)
    ds_by_cam[cam] = ds
```
