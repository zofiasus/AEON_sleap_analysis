# .slp file to video .mp4 conversion with inference points

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path("/Users/zosiasus/Documents/Aeon3_SLEAP/AEON_sleap_analysis")
SCRIPTS_DIR = PROJECT_ROOT / "scripts"

if str(SCRIPTS_DIR) not in sys.path:
    sys.path.append(str(SCRIPTS_DIR))

from sleap_inference_report import ingest_data, export_video_large, export_video_small
```

```python
predictions_dir = Path(
    "/Users/zosiasus/Documents/Aeon3_SLEAP/aeon3_data/abcEphysPilot01_sleap_for_card/models/card_inference_results/predictions"
)
north_slp_path = next(predictions_dir.glob("CameraNorth*.slp"))
south_slp_path = next(predictions_dir.glob("CameraSouth*.slp"))
west_slp_path = next(predictions_dir.glob("CameraWest*.slp"))
east_slp_path = next(predictions_dir.glob("CameraEast*.slp"))
top_slp_path = next(predictions_dir.glob("CameraTop*.slp"))
nest_slp_path = next(predictions_dir.glob("CameraNest*.slp"))
```

```python
north_data = ingest_data(next(predictions_dir.glob("CameraNorth*.slp")))
south_data = ingest_data(next(predictions_dir.glob("CameraSouth*.slp")))
west_data = ingest_data(next(predictions_dir.glob("CameraWest*.slp")))
east_data = ingest_data(next(predictions_dir.glob("CameraEast*.slp")))
top_data = ingest_data(next(predictions_dir.glob("CameraTop*.slp")))
nest_data = ingest_data(next(predictions_dir.glob("CameraNest*.slp")))
```
